```python
import sys; sys.path.insert(0, "../..")
import os
import sys
import subprocess
from glob import glob
from os import path

from pyrosetta import *
import pyrosetta.toolbox as tb
import ssOpen
pyrosetta.init()


#folder=sys.argv[1]
folder='.'

pdb_files =  glob(path.join(folder, "struct/*.pdb"))
ss_files = [file+".ss2" for file in pdb_files]
ss2=True
```

```python
count=0

for x in range(0,len(pdb_files)):

    pose = pyrosetta.rosetta.core.import_pose.pose_from_file(pdb_files[x])
    tb.get_secstruct(pose, output=False)
    ss = pose.secstruct()

    with open(ss_files[x]) as f:
        data = f.readlines()
        if(ss2):
            data=data[2:]
        
        ssObj = ssOpen.ss(data,ss_files[x])
    
    print("SEQ : "+pose.sequence())
    print("POSE: "+ss)
    print("PSIP: "+ssObj.ss)
    print()
```
